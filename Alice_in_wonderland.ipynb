```python
from wordcloud import WordCloud,STOPWORDS
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import nltk
```

```python
path='./Data/alice_novel.txt'
alice_novel=open(path,'r').read()
```

```python
alice_novel
```

```python
stopwords=set(STOPWORDS)
```

```python
stopwords
```

```python
stopwords.add('said')
alice_wc=WordCloud(
    background_color='white',
    stopwords=stopwords,
    max_words=25
)
alice_wc.generate(alice_novel)
plt.imshow(alice_wc)
plt.axis('off')
plt.show()
```

```python
nltk.download('punkit')
nltk.download('averaged_perceptron_tagger')
```

```python
words=nltk.word_tokenize(alice_novel)
postag=nltk.pos_tag(words)

noun=[]
verb=[]
adjectives=[]

for word,each_p in postag:
    if each_p in ['NN','NNS','NNP','NNPS']:
        noun.append(word)
    if each_p in ['VB','VBD','VBG','VBN','VBP','VBZ']:
        verb.append(word)
    if each_p in ['JJ','JJR','JJS']:
         adjectives.append(word)

```

```python
noun_text=" ".join(noun)
verb_text=" ".join(verb)
adjectives_text=" ".join(adjectives)
```

```python
alice_noun_text_wc=WordCloud(
    background_color='white',
    stopwords=stopwords,
    max_words=30
)
alice_noun_text_wc.generate(noun_text)
plt.imshow(alice_noun_text_wc)
plt.axis('off')
plt.show()
```

```python
alice_verb_text_wc=WordCloud(
    background_color='white',
    stopwords=stopwords,
    max_words=30
)
alice_verb_text_wc.generate(verb_text)
plt.imshow(alice_verb_text_wc)
plt.axis('off')
plt.show()
```

```python
alice_adjectives_text_wc=WordCloud(
    background_color='white',
    stopwords=stopwords,
    max_words=30
)
alice_adjectives_text_wc.generate(adjectives_text)
plt.imshow(alice_adjectives_text_wc)
plt.axis('off')
plt.show()
```
