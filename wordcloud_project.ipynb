```python
from wordcloud import WordCloud,STOPWORDS
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import nltk
```

```python
path='./Data/a_new_hope.txt'
hope_data=open(path,'r').read()
```

```python
stopwords=set(STOPWORDS)
```

```python
words=nltk.word_tokenize(hope_data)
postag=nltk.pos_tag(words)

noun=[]
verb=[]
adjectives=[]

for word,each_p in postag:
    if each_p in ['NN','NNS','NNP','NNPS']:
        noun.append(word)
    if each_p in ['VB','VBD','VBG','VBN','VBP','VBZ']:
        verb.append(word)
    if each_p in ['JJ','JJR','JJS']:
         adjectives.append(word)
```

```python
noun_text=" ".join(noun)
verb_text=" ".join(verb)
adjective_text=" ".join(adjectives)

```

```python
noun_text_wc=WordCloud(
    background_color='white',
    stopwords=stopwords,
    max_words=200,
    max_font_size=40,
    random_state=42
)

noun_text_wc.generate(noun_text)
plt.figure(figsize=(16,10))
plt.imshow(noun_text_wc,interpolation='bilinear')
plt.axis('off')

plt.show()
```

```python
verb_text_wc=WordCloud(
    background_color='white',
    stopwords=stopwords,
    max_words=200,
    max_font_size=40,
    random_state=42
)

verb_text_wc.generate(verb_text)
plt.figure(figsize=(16,10))
plt.imshow(verb_text_wc,interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
adjective_text_wc=WordCloud(
    background_color='white',
    stopwords=stopwords,
    max_words=200,
    max_font_size=40,
    random_state=42
)

adjective_text_wc.generate(adjective_text)
plt.figure(figsize=(16,10))
plt.imshow(adjective_text_wc,interpolation='bilinear')
plt.axis('off')
plt.show()
```
